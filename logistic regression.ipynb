```python
#Function to scale the variable for all feature, so the weights as not too large or too small
def scale_dataset(dataframe):
    X = dataframe[dataframe.columns[2:32]].values
    y = dataframe[dataframe.columns[1]].values
    
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    
    data = np.hstack((X , np.reshape(y, (-1,1))))
    
    #features that are removed from the model
    remove_var = [1,4,8,9,11,14,15,16,17,18,19,21,24]  
    count = 0
    for i in remove_var:
        data = np.delete(data, (i - count), 1)
        X = np.delete(X, (i - count), 1)
        count += 1
    
    return data, X, y
```

```python
#Calculating 1/(1 + e^-z) for the sigmoid function for logistic regression
def sigmoid(z):
    
    # input Z = w*x + b
    
    # output g = sigmoid function
    
    g = 1/(1 + np.exp(-z))
    
    return g
    
```

```python
#Calculating cost for testing and for gradient descent algorithm
def cost_function(X,y,w,b,lambda_):
    
    # input 
    # X = training set
    # y = true value
    # w = weights
    # b = Bias 
    # lambda_ = regularization variable
    
    # output 
    # total_cost = how the model is performing
    
    m,n = X.shape

    cost = 0.0
    reg_cost_j = 0.0
    reg_cost = 0.0
    
    for i in range(m):
        z = np.dot(X[i],w) + b
        f_wb = sigmoid(z)
        cost = cost - y[i]*np.log(f_wb) - (1-y[i])*np.log(1 - f_wb)
    
    # regularization variable so the weights as not too large or too small
    for j in range(n):
        reg_cost_j = w[j]**2
        reg_cost = reg_cost + reg_cost_j
        
    reg_cost = (lambda_/(2 * m)) * reg_cost  
    
    total_cost = cost + reg_cost
    total_cost = total_cost / m
    
    return total_cost
        
```

```python
#Calculating Gradient for the gradient descent 
def compute_gradient(X,y,w,b,lambda_):
    
    # input 
    # X = training set
    # y = true value
    # w = weights
    # b = Bias 
    # lambda_ = regularization variable
    
    # output 
    # dj_dw, dj_db = Gradient with respect to W and b 
    
    m,n = X.shape
    dj_dw = np.zeros(w.shape)
    dj_db = 0.
    
    for i in range(m):
        z = np.dot(X[i],w) + b
        f_wb = sigmoid(z) 
        dj_dw_i = f_wb - y[i] 
        for j in range(n):
            dj_dw[j] = dj_dw[j] + dj_dw_i * X[i,j] + (lambda_ / m) * w[j]
            
    dj_db =  dj_db + dj_dw_i
    
    
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    
    return dj_dw, dj_db
```

```python
#Function calculates the gradient and updates the weights w and b.
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters, lambda_):
    
    # input 
    # X = training set
    # y = true value
    # w_in = weights
    # b_in = Bias 
    # cost_function = to calculate the cost
    # gradient_function = to calculate the gradient
    # alpha = step-size to update weights and bias
    # num_iters = no, of iterations
    # lambda_ = regularization variable
    
    # output 
    # w_in = weights
    # b_in = bias
    # J_history = variable to store past cost results
    # w_history = variable to store past weights
    
    J_history = []
    w_history = []
    
    for i in range(num_iters):
        
        dj_dw , dj_db = gradient_function(X,y,w_in,b_in,lambda_)
        
        w_in = w_in - alpha * dj_dw 
        
        b_in = b_in - alpha * dj_db
    
    if i<100000:      # prevent resource exhaustion 
        cost =  cost_function(X, y, w_in, b_in,lambda_)
        J_history.append(cost)

    # Print cost every at intervals 10 times or as many iterations if < 10
    if i% math.ceil(num_iters/1) == 0 or i == (num_iters-1):
        w_history.append(w_in)
        print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
        
    return w_in, b_in, J_history, w_history
```

```python
np.random.seed(1)
initial_w =  np.random.rand(n) - 0.5
initial_b = 1

iterations = 1000
alpha = 0.1

w,b, J_history,_ = gradient_descent(X_train ,y_train, initial_w, initial_b, 
                                   cost_function, compute_gradient, alpha, iterations, lambda_)
print(w,b)
```

```python
#To check the accurcay of the model
def predict(X,w,b):
    m, n = X.shape   
    p = np.zeros(m)

    for i in range(m):   
        z_wb = 0.0

        for j in range(n): 
            z_wb_ij = X[i, j] * w[j]
            z_wb += z_wb_ij
        z_wb += b
        f_wb = sigmoid(z_wb)
        #print(f_wb)

        p[i] = f_wb >= 0.5

    return p
```
